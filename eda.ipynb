```python
import pandas as pd

df = pd.read_csv("cleaned_titanic.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
df.hist(figsize=(10,8))
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x=df['Fare'])
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df)
plt.show()
```

## Key Observations:

- Passengers in higher class (Pclass) had higher survival rate
- Fare shows presence of outliers, indicating some passengers paid significantly higher
- Age distribution is slightly skewed with most passengers in young-adult range
- Positive relationship between Fare and Survival can be observed
- Correlation between most features is weak, indicating independent variables

## Conclusion:

Exploratory Data Analysis helped in understanding the dataset, identifying patterns, detecting outliers, and recognizing important features that can impact machine learning models.
